## 11. Mtodos e Funes de Listas

```python
# MTODOS DE LISTAS  uma funo que est atrelado a uma variavl.

lista = [1, 3, 12, 8, 2]

# append - adiciona elementos a lista sempre no final da lista 
print(f'Lista antes do append \n {lista}')
lista.append(3)
print(f'Lista depois do append \n {lista}')

# insert adiciona os elemento na lista em qualquer lugar da lista
lista.insert(2,10)
print(f'Lista depois do insert \n {lista}')

#Extend  junta as listas, porem a lista2 vai para o final 
lista2 = [1, 2, 3]
lista.extend(lista2)
print(f'Lista depois do extend \n {lista}')

# pop remover elementos 
lista.pop() # remove o ultimo elemento da lista 
print(f'Lista depois do pop \n {lista}')

lista.pop(0) # remove o ndice 0 da lista
print(f'Lista depois do pop(0) \n {lista}')

# remove - remove o elemento da lista
lista.remove(3)
print(f'Lista depois do pop(0) \n {lista}')

# count - conta quantos elementos 2 tem na lista
print(f'Quantidade de 2 na lista:{lista.count(2)}')

# index mostra o elemento do ndice. 
print(f'ndice do elemento 12:', lista.index(12))

# sort ordena a lista 
lista.sort(reverse=True)
print(f'Lista depois do sort(reverse=True) \n {lista}')
lista.sort()
print(f'Lista depois do sort(0) \n {lista}')

# Funes de lista 

#len 
print(f'Funes de lista len \n {len(lista)}')

# sum
print(f'Funes de lista sum  \n {sum(lista)}')

# Max
print(f'Funes de lista max Mxima lista  \n {max(lista)}')

#min
print(f'Funes de lista min - Mnimo lista \n {min(lista)}')

```

```python
# 2. Criao de funes 

# Funo inicial com parametros 

def saudacao(name):
    print('Seja bem vindo' )
    print(f'ola,  um prazer ter voc fazendo parte desse curso, {name}')
saudacao("Givanildo ")

print('---------------------------- \n') 

# Funo com parmetros 
def saudacao2(nome, curso='Python'):
    print(f'Seja bem vindo, {nome}' )
    print(f'ola,  um prazer ter voc fazendo parte desse curso {curso}')
saudacao2("Givanildo","Python")

print('---------------------------- \n') 

# Funes com retorno

def soma(num1,num2):
    return num1+num2

resultado = soma(5,7)
print(f'O resultado da soma  {resultado}')
    

```

```python
def calculadora(num1, num2, operao = '+'):
    if operao == '+':
        return num1 + num2
    elif operao == '-':
        return num1-num2
    elif operao == '*':
        return num1 *num2
    else:
        return num1/num2
    
    
resultado = calculadora(4, 2, "+")
print(resultado)


```
