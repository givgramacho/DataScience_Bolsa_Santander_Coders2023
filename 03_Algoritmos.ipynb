```python
print("--- Moisés, bem-vindo ao App Ada Poupança ---")

saldo = 1000
poupanca = 500

i = 0

while i != 3:
    i = int(input("Digite 1 para APLICAR, 2 para RESGATAR e 3 para SAIR: \n"))
    if i == 1:
        valor = float(input("Valor da aplicação: \n"))
        if valor > saldo:
            print("Saldo de conta corrente insuficiente \n")
        else:
            saldo = poupanca + valor
            poupanca = poupanca +valor
            print("Aplicação realizada com sucesso \n")
    else:
        if i == 2:
            valor = float(input("Valor do Resgate: \n "))
            if valor > poupanca:
                print("Saldo de poupança insuficiente \n")
            else:
                saldo = saldo + valor
                poupanca = poupanca - valor
                print("Resgate realizado com sucesso \n")
        else:
            print("Até logo!")

            
```
